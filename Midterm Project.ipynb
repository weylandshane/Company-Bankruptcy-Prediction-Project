```python
#It appears we are only dealing with float and int data, no chars nor strings involved
#the first column seems to be useless, and net income flag seems to always be 1, and the Liability-Assets Flag seems to almost 
#always be 0 and carries no information, so we go ahead and delete them
df.drop('Unnamed: 0', axis = 1, inplace=True)
df.drop(' Net Income Flag',axis = 1, inplace=True)
df.drop(' Liability-Assets Flag',axis = 1, inplace=True)
```

```python
#we are working on bankruptcy data, so we wanna investigate it first

#using code from lecture "imbalanced data"
target_count = df['Bankrupt?'].value_counts()
print('Class 0:', target_count[0])
print('Class 1:', target_count[1])
print('Proportion:', round(target_count[0] / target_count[1], 2), ': 1')

target_count.plot(kind='bar', title='Count (target)');
```

```python
#170 out of the 5454 entries of data are actually bankrupted, so keep in mind that we are dealing with highly unbalanced data here. 
#We are gonna choose oversampling to help with the unbalanced data
```

```python
#after investigating the response variable, we wanna research the explanatory variables.
corr = df.corr()
corr
```

```python
#ROA(A),ROA(B),ROA(C) in the top left corner seems to be highly correlated, so we only keep one of the three
#Since we are dropping ROA of A and B, it makes sense to also drop Net Value per share of A and B
df.drop(' ROA(A) before interest and % after tax',axis = 1, inplace=True)
df.drop(' ROA(B) before interest and depreciation after tax',axis = 1, inplace=True)
df.drop(' Net Value Per Share (A)',axis = 1, inplace=True)
df.drop(' Net Value Per Share (B)',axis = 1, inplace=True)
```

```python
#some of the variables show very low correlation to the independent varabile, so we wanna delete those variables
corr = df.corr()
```

```python
#finding the redundant variables with high correlation
```

```python
cols = [" Current Liabilities/Equity"," Current Liabilities/Liability"," Debt ratio %"," Operating Gross Margin",
        " Realized Sales Gross Margin"," Operating Profit Per Share (Yuan ¥)"," After-tax Net Profit Growth Rate",
       " Pre-tax net Interest Rate"," Persistent EPS in the Last Four Seasons"," After-tax net Interest Rate",
       " Pre-tax net Interest Rate"," Net Value Growth Rate"," After-tax net Interest Rate"]
cols
```

```python
#drop the columns with high correlation with each other
df.drop(cols,axis = 1, inplace=True)
#check the new correlation matrix
corr = df.corr()
corr
```

```python
#start oversampling our imbalanced data
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,accuracy_score
from sklearn.linear_model import LogisticRegression

from imblearn import over_sampling
from imblearn.over_sampling import SMOTE

# Remove 'Bankrupt?' column
labels = df.columns[:-1]
labels

x = df[labels]
y = df['Bankrupt?']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

#oversampling x_train and y_train data
concat = pd.concat([x_train, y_train], axis=1)
X = concat[labels]
Y = concat['Bankrupt?']
sm = SMOTE(random_state=15)
X_res, Y_res = sm.fit_resample(X, Y)

#check the distribution of the oversampled training data
Y_res.value_counts()
```

```python
#Method 1: lOGISTIC REGRESSION

#train the model using resampled data 
lr = LogisticRegression()
lr.fit(X_res,Y_res)

#use the trained model to make predictions on validation set
logpredict = lr.predict(x_test)

#report accuracy
print("Result: trained on the oversampled training data")
print(confusion_matrix(y_test,logpredict))
print(classification_report(y_test,logpredict))
print("accuracy: ",accuracy_score(y_test, logpredict))
print("auc score: ",roc_auc_score(y_test, logpredict))
print()




#train the model using unresampled data 
lr.fit(x_train,y_train)

#use the trained model to make predictions on validation
logpredict = lr.predict(x_test)

#report accuracy
print("Result: trained on the non-oversampled training data")
print(confusion_matrix(y_test,logpredict))
print(classification_report(y_test,logpredict))
print("accuracy: ",accuracy_score(y_test, logpredict))
print("auc score: ",roc_auc_score(y_test, logpredict))

print()
```

```python
#Result :
#we conclude that oversamling method does help with the accuracy of the logModel on the class_1, 
#despite them having a negative impact on the class_0 case and overall accuracy
#in the following test we are only focusing on the oversampled data
#Overall logistic regression is not a very good method
```

```python
#Method 2: KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing

#scaling the variables
scaler=preprocessing.StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X_res))
```

```python
#find the n with the highest auc value
import numpy as np
from matplotlib import pyplot as plt

auc = []
for i in range(1,50):
 knn = KNeighborsClassifier(n_neighbors=i)
 knn.fit(X,Y_res)
 pred = knn.predict(x_test)
 auc.append(roc_auc_score(y_test, pred))

```

```python
max_val = np.max(auc)
val = auc.index(max_val)+1
val
```

```python
#when n=val , knn has largest auc score
```

```python
knn_model = KNeighborsClassifier(n_neighbors=val)
knn_model.fit(X, Y_res)
predict = knn_model.predict(x_test).round()

#report accuracy
print("Result: trained on the oversampled training data")
print(confusion_matrix(y_test,predict))
print(classification_report(y_test,predict))
print("accuracy: ",accuracy_score(y_test, predict))
print("auc score: ",roc_auc_score(y_test, predict))
print()
```

```python
#We can see that the knn method , so it is not the best method
```

```python
#Method 3: RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier()

#using code from "Decision tree" lecture
param_grid = {
    'n_estimators': [50,60,70,80,90,100,110,120,130,140,150]
}
grid_searcher = GridSearchCV(estimator=rf, param_grid=param_grid, n_jobs=-1)
grid_searcher.fit(X_res, Y_res)
print(grid_searcher.best_score_)
print(grid_searcher.best_params_)

num = grid_searcher.best_params_['n_estimators']
```

```python
rf = RandomForestClassifier(num)
rf.fit(X_res,Y_res)
predict1 = rf.predict(x_test)
```

```python
#report accuracy
print("Result: trained on the oversampled training data")
print(confusion_matrix(y_test,predict1))
print(classification_report(y_test,predict1))
print("accuracy: ",accuracy_score(y_test, predict1))
print("auc score: ",roc_auc_score(y_test, predict1))
print()
```

```python
#Random forest appears to be the most accurate method
```
